Simple LLM Base sequential question answer workflow

```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
 
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from dotenv import load_dotenv
import os
```

```python
# Your Hugging Face Key
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "HuggingFace_API_Key"

# Verify the Hugging Face token is loaded
print(str(os.environ.get("HUGGINGFACEHUB_API_TOKEN"))[:10] + "...")
```

```python
# Step 1: create the endpoint with conversational task
endpoint = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen2.5-7B-Instruct",
    task="conversational",
    max_new_tokens=256,
)

llm = ChatHuggingFace(llm=endpoint)
```

```python
# State
class LLMState(TypedDict):
    question: str
    ans: str

# Node — now result.content works correctly because ChatHuggingFace returns a message
def LLM_Q(state: LLMState) -> LLMState:
    question = state["question"]
    prompt = f"Answer this question wisely: {question}"
    result = llm.invoke(prompt)
    state["ans"] = result.content   # ✅ .content works with ChatHuggingFace
    return state
```

```python
# Build graph
graph = StateGraph(LLMState)
graph.add_node("LLM_Q", LLM_Q)
graph.add_edge(START, "LLM_Q")
graph.add_edge("LLM_Q", END)
workflow = graph.compile()
```

```python
# Run
initial_state = {"question": "Who is the owner of Tesla and Google?", "ans": ""}
final_state = workflow.invoke(initial_state)
print(final_state["ans"])
```
